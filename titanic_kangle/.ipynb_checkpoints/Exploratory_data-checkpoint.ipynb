# 4. Feature Engineering & Data Wrangling

# 1 Title

```python
import re

def extract_title(name):
    p = re.compile(r',([\w\s]+)\.')
    return p.search(name).groups(1)[0].strip()

dataset['Name'].apply(lambda name: extract_title(name))


dataset['Title'] = dataset['Name'].apply(lambda name: extract_title(name))
test_df['Title'] = dataset['Name'].apply(lambda name: extract_title(name))


def group_title(title):
    if title in ['Mr', 'Miss', 'Mrs', 'Master']:
        return title
    elif title == 'Ms':
        return 'Miss'
    else:
        return  'others'
            

dataset['Title'] = dataset['Title'].apply(lambda title: group_title(title))
```

```python
test_df.columns
```

```python
test_df['Title'] = test_df['Title'].apply(lambda title: group_title(title))
```

```python
sns.countplot(data = dataset, x = 'Title', hue = 'Survived')
```

## 4.2 Family

```python
dataset["Family_Size"] = dataset['SibSp'].astype('int') + dataset['Parch'].astype('int') + 1
test_df["Family_Size"] = test_df['SibSp'].astype('int') + test_df['Parch'].astype('int') + 1
sns.countplot(data = dataset, x = 'Family_Size', hue = 'Survived')
```
